Base score will be based on the direction and value of the 1-Day Moving Average

```python
### This is set to 10 days. See below cells for 7 days.

def base_score_10_day(symbol, name):
    try:
        price_df = pandas.read_csv(os.getcwd() + "\\Daily Stock Prices\\" + symbol + ' - ' + name + '.csv', index_col= 'Date', parse_dates= True)
        price_df = price_df.sort_index(ascending=True)

        
        print(name)

        counter = 0
        combined_df = pandas.DataFrame()
        for i in range(0, len(price_df)):
            try:
                if i < 10:
                    temp_df = pandas.DataFrame({'Symbol': symbol, 'Name': name, 'Date': price_df.index[i], 'Base_Score': numpy.NaN, 'Trend': numpy.NaN, 'Average_10_Day_Price': numpy.NaN, 'Average_10_Day_Slope': numpy.NaN}, index=[counter])

                    combined_df = pandas.concat([combined_df, temp_df])

                    counter += 1

                else:
                    spliced_price_df = price_df.iloc[i-9:i+1]
                    ## Calculating Simple Moving Average, 1 Days:
                    SMA_2 = spliced_price_df.Close.rolling(window=1).mean()

                    ## Slope of between the previous 10 days is the trend/value evaluater:
                    trend_list = []

                    for x in range(len(SMA_2)):
                        trend_list.append(SMA_2[x])

                    slope_list = []

                    for x in range(len(SMA_2)-1):
                        start_step = trend_list[x]
                        end_step = trend_list[x+1]
                        slope = end_step - start_step
                        slope_list.append(slope)

                    average_slope = numpy.nanmean(slope_list)
                    average_price = numpy.nanmean(trend_list)

                    stock_score = 0
                    trend = ''

                    if average_price >= 30:
                        if average_slope >= .50:
                            stock_score = 50
                            trend = 'Slight Uptrend'
                            if average_slope >= .85 :
                                stock_score = 100
                                trend = 'Uptrend'
                                if average_slope >= 1.25:
                                    stock_score = 150
                                    trend = 'High Uptrend'
                        elif average_slope <= -.50:
                            stock_score = -50
                            trend = 'Slight Downtrend'
                            if average_slope <= -.85 :
                                stock_score = -100
                                trend = 'Downtrend'
                                if average_slope <= -1.25:
                                    stock_score = -150
                                    trend = 'High Downtrend'
                        else:
                            stock_score = 0
                            trend = 'No Trend'

                    elif average_price <= 30:
                        if average_slope >= .30:
                            stock_score = 50
                            trend = 'Slight Uptrend'
                            if average_slope >= .50 :
                                stock_score = 100
                                trend = 'Uptrend'
                                if average_slope >= .75:
                                    stock_score = 150
                                    trend = 'High Uptrend'
                        elif average_slope <= -.30:
                            stock_score = -50
                            trend = 'Slight Downtrend'
                            if average_slope <= -.50 :
                                stock_score = -100
                                trend = 'Downtrend'
                                if average_slope <= -.75:
                                    stock_score = -150
                                    trend = 'High Downtrend'
                        else:
                            stock_score = 0
                            trend = 'No Trend'


                    temp_df = pandas.DataFrame({'Symbol': symbol, 'Name': name, 'Date': spliced_price_df.index[-1], 'Base_Score': stock_score, 'Trend': trend, 'Average_10_Day_Price': average_price, 'Average_10_Day_Slope': average_slope}, index=[counter])

                    combined_df = pandas.concat([combined_df, temp_df])

                    counter += 1
                
            except Exception:
                print('Error 10')
                pass

        if not os.path.exists(os.getcwd() + '\\Base Score\\'):
            os.makedirs(os.getcwd() + '\\Base Score\\')
            
        combined_df.to_csv(os.getcwd() + '\\Base Score\\'  + symbol + ' - ' + name + ' - Base Score - 10 Day.csv')

        print(10)
        
        return combined_df

    except Exception:
        pass
```

```python
### Looping:
stocks_and_names_with_indices = pandas.read_csv('merged_NYSE_AMEX_removed_intercept_pattern.csv')

for x, y in stocks_and_names_with_indices.iterrows():
    try:
        price_df = pandas.read_csv(os.getcwd() + "\\Daily Stock Prices\\" + y['Symbol'] + ' - ' + y['Description'] + '.csv', index_col= 'Date', parse_dates= True).sort_index(ascending=True)
    
    except Exception:
        continue
        
    try:
        base_score_10_day_df = base_score_10_day(y['Symbol'], y['Description'])
        base_score_7_day_df = base_score_7_day(y['Symbol'], y['Description'])
        base_score_5_day_df = base_score_5_day(y['Symbol'], y['Description'])

        ### Final Base Score = addition of all base scores from 5, 7, and 10 day ranges:
        ### Price Estimator = (average of all the slopes * 5 days) + current price
        Date_list = []
        base_summary_df = pandas.DataFrame()
        index_counter = 0

        for u, v in base_score_5_day_df.iterrows():
            splice_10_day_slope = numpy.NaN
            splice_7_day_slope = numpy.NaN
            splice_5_day_slope = v.Average_5_Day_Slope
            combine_trend = v.Trend

            try:
                splice_10_day_slope = base_score_10_day_df.Average_10_Day_Slope[base_score_10_day_df.Date == v.Date]
                if len(splice_10_day_slope) == 0:
                    splice_10_day_slope = numpy.NaN
            except:
                pass

            try:
                splice_7_day_slope = base_score_7_day_df.Average_7_Day_Slope[base_score_7_day_df.Date == v.Date]
                if len(splice_7_day_slope) == 0:
                    splice_7_day_slope = numpy.NaN
            except:
                pass

            average_slope = numpy.nanmean([splice_5_day_slope, float(splice_7_day_slope), float(splice_10_day_slope)])

            predicted_value_for_end_date = average_slope * 5 + float(price_df.Close[price_df.index == v.Date.to_pydatetime()]) # 5 days, linear equation

            if price_df.index.get_loc(v.Date.to_pydatetime()) + 5 < len(price_df):        
                actual_value_for_end_date = float(price_df.iloc[price_df.index.get_loc(v.Date.to_pydatetime())+5]['Close'])

                predicted_actual_difference = actual_value_for_end_date - predicted_value_for_end_date

            else:
                actual_value_for_end_date = numpy.NaN

                predicted_actual_difference = numpy.NaN

            if v.Trend == 'No Trend':
                try:
                    if base_score_10_day_df.Trend != 'No Trend':
                        combine_trend = base_score_10_day_df.Trend
                except Exception:
                    pass

                try:
                    if base_score_7_day_df.Trend != 'No Trend':
                        combine_trend = base_score_7_day_df.Trend
                except Exception:
                    pass


            temp_df = pandas.DataFrame({'Name': y['Description'], 'Symbol': y['Symbol'], 'Date': v.Date, 'Trend': combine_trend, 'Average_Slope': average_slope, 'Predicted_Value_For_End_Date': predicted_value_for_end_date, 'Actual_Value_At_End_Date': actual_value_for_end_date, 'Value_Difference': predicted_actual_difference}, index=[index_counter])

            index_counter += 1

            base_summary_df = pandas.concat([base_summary_df, temp_df])

        base_summary_df.to_csv(os.getcwd() + '\\Base Score\\'  + y['Symbol'] + ' - ' + y['Description'] + ' - Base Score - Summary.csv')
        
    except Exception:
        print('Error: ', y['Description'])
        pass
```
